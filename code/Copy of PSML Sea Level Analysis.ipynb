```python
import plotly.graph_objects as go

# Define the coordinates for the Gulf of Mexico
gulf_of_mexico_lon = [-100, -80, -80, -100, -100]
gulf_of_mexico_lat = [15, 15, 30, 30, 15]

# Define the coordinates for the North Atlantic Ocean
north_atlantic_lon = [-80, -20, -20, -80, -80]
north_atlantic_lat = [0, 0, 80, 80, 0]

# Define the coordinates for the North Pacific Ocean
north_pacific_lon = [-180, -100, -100, -180, -180]
north_pacific_lat = [0, 0, 60, 60, 0]

# Define the coordinates for the Bering Sea
bering_sea_lon = [-180, -160, -160, -180, -180]
bering_sea_lat = [50, 50, 60, 60, 50]

# Create the 3D globe plot
fig = go.Figure()

# Add a scattergeo trace for your existing points
fig.add_trace(go.Scattergeo(lat=[40.7607, 51.4934], lon=[-111.891, 0.0098], mode='markers', marker_size=20, line=dict(color='Red')))

# Add a shape trace to mark the North Atlantic Ocean
fig.add_trace(go.Scattergeo(lat=north_atlantic_lat, lon=north_atlantic_lon, mode='lines', line=dict(color='blue', width=2)))

# Add a shape trace to mark the Gulf of Mexico
fig.add_trace(go.Scattergeo(lat=gulf_of_mexico_lat, lon=gulf_of_mexico_lon, mode='lines', line=dict(color='green', width=2)))

# Add a shape trace to mark the North Pacific Ocean
fig.add_trace(go.Scattergeo(lat=north_pacific_lat, lon=north_pacific_lon, mode='lines', line=dict(color='orange', width=2)))

# Add a shape trace to mark the Bering Sea
fig.add_trace(go.Scattergeo(lat=bering_sea_lat, lon=bering_sea_lon, mode='lines', line=dict(color='purple', width=2)))

# Set the projection type to orthographic
fig.update_geos(projection_type="orthographic")

# Adjust layout
fig.update_layout(width=800, height=800, margin={"r": 0, "t": 0, "l": 0, "b": 0})

# Save the plot as HTML
fig.write_html("3d_plot_with_regions.html")

# Show the plot
fig.show()

```
